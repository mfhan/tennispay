```python
def get_tennis_data(url):
    response = requests.get(url)
    response.raise_for_status()  # Check if request was successful
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extracting the update date
    update_text = soup.find('div', {'class': 'h2h_text'}).text
    update_date_str = update_text.split(': ')[1]
    update_date = datetime.strptime(update_date_str, '%Y.%m.%d %H:%M:%S').strftime('%Y-%m-%d')

    # Locate the table
    table = soup.find('table', {'id': 'u20'})

    # Locate rows
    rows = table.find('tbody').find_all('tr')

    data = []
    for row in rows[:50]:  # Limit to top 50 players
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        player_data = [ele for ele in cols if ele]
        player_data.insert(0, update_date)  # Insert the update date at the beginning of player data
        data.append(player_data)

    return data
```

```python
url = 'https://live-tennis.eu/en/atp-live-ranking'
get_tennis_data(url)
```
