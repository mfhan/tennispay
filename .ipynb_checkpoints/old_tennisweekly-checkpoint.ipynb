```python
# Create a CSV file to store the data
with open("atp_live_ranking.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row
    writer.writerow(["Rank", "Player", "Points"])

    # Iterate over the table rows and write the data to the CSV file
    if table is not None:
        for row in table.find_all("tr"):
            columns = row.find_all("td")
    
            # Get the rank, player, and points
            rank = columns[0].text
            player = columns[1].text
            points = columns[2].text
    
            # Write the data to the CSV file
            writer.writerow([rank, player, points])

# Close the CSV file
csvfile.close()
```

```python
soup = bs4.BeautifulSoup(URL, "html.parser")

# Find the top 100 players
players = soup.find_all("div", class_="row-player")

# Create a CSV file to store the data
with open("atp_live_ranking.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row
    writer.writerow(["Player", "Country Code", "Full Name", "Age", "Point Total"])

    # Iterate over the players and write the data to the CSV file
    for player in players:
        # Get the player's country code
        country_code = player.find("div", class_="spr").text

        # Get the player's full name
        full_name = player.find("div", class_="pn").text

        # Get the player's age
        age = player.find("div", class_="pa").text

        # Get the player's point total
        point_total = player.find("div", class_="pts").text

        # Write the data to the CSV file
        writer.writerow([full_name, country_code, age, point_total])

# Close the CSV file
csvfile.close()
```
