```python
from playwright.async_api import async_playwright
import pandas as pd
from bs4 import BeautifulSoup 
import requests
import datetime
import time
import json
import csv
import numpy as np
import re
import lxml.html
```

```python
# Set the URL of the ATP live ranking page
URL = "https://live-tennis.eu/en/atp-live-ranking"

# Create a Playwright sync context
playwright = await async_playwright().start()
```

```python
# Create a Playwright browser
browser = await playwright.chromium.launch(headless=False)
```

```python
# Create a Playwright page
page = await browser.new_page()
```

```python
#time.sleep(10)
page.wait_for_timeout(5000)
```

```python
# Navigate to the ATP live ranking page
await page.goto(URL)
```

```python
# Wait for the page to load
page.wait_for_load_state("networkidle")
```

```python
# Get the page content
html = await page.content()
```

```python
# Close the browser
browser.close()
```

```python
# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")
```

```python
table = soup.find("table", id='u868')
```

```python
body = table.find_all('tbody')
```

```python
print(body[12])
```

```python
players = body[12].find_all('tr')
```

```python
top_players = players[0:16]
```

```python
print(top_players)
```

```python
# define current date 
from datetime import datetime
current_date = datetime.now().strftime('%Y-%m-%d')
```

```python
data = {
    "rank": [],
    "career high": [],
    "name": [],
    "age": [],
    "country": [],
    "points": [],
    "next event": [],
    "next points": [],
    "max points": []
}

#age = int(player.find().text)

for player in top_players:
    # soup = BeautifulSoup(row, 'html.parser')   
    rank = player.find('td', class_='rk').text if player.find('td', class_='rk') else ""
    career_high = player.find('td', class_='chtd').text if player.find('td', class_='chtd') else ""
    name = player.find('td', class_='pn').text if player.find('td', class_='pn') else ""
    age = player.find('td', class_='sm') if player.find('td', class_='sm') else ""
    country = player.find('td', class_='sm', p="1").text if player.find('td', class_='sm', p="1") else ""
    all_td = player.find_all('td') 
    points = all_td[6] if len(all_td)> 6 else ""
    next_event = player.find('td', class_='rst tc').text if player.find('td', class_='rst tc') else ""
    next_points = all_td[11] if len(all_td)> 11 else ""
    # next_int_points = int(next_points) if len(next_points) > 0 else 0
    max_points = all_td[12] if len(all_td)> 12 else ""
    # max_int_points = int(max_points) if len(max_points) > 0  else 0
    # next_points = player.find('td', class_='hdc').text if player.find('td', class_='hdc') else ""
    # max_points = player.find_all('td', class_='hdc')[-1].text if player.find_all('td', class_='hdc') else ""

    # Append data to lists
    data["rank"].append(rank)
    data["career high"].append(career_high)
    data["name"].append(name)
    data["age"].append(age)
    data["country"].append(country)
    data["points"].append(points)
    data["next event"].append(next_event)
    data["next points"].append(next_points)
    data["max points"].append(max_points)
    # Create the DataFrame
df = pd.DataFrame(data)
df["date"] = current_date


# Display the DataFrame
#print(df)
df.head(10)
#df.shape
# df.tail()
df.to_csv('weekly_ranks.csv')
```

```python
data = {
    "rank": [],
    "name": [],
    "age": [],
    "country": [],
    "points": []
}

for player in finalists:
    # soup = BeautifulSoup(row, 'html.parser')   
    rank = player.find('td', class_='rk').text if player.find('td', class_='rk') else ""
    name = player.find('td', class_='pn').text if player.find('td', class_='pn') else ""
    age = player.find('td', class_='sm') if player.find('td', class_='sm') else ""
    country = player.find('td', class_='sm', p="1").text if player.find('td', class_='sm', p="1") else ""
    all_td = player.find_all('td') 
    points = all_td[6] if len(all_td)> 6 else ""


    # Append data to lists
    data["rank"].append(rank)
    data["name"].append(name)
    data["age"].append(age)
    data["country"].append(country)
    data["points"].append(points)
    # Create the DataFrame
race_df = pd.DataFrame(data)
race_df["date"] = current_date


# Display the DataFrame
#print(df)
race_df.head(10)
#df.shape
# df.tail()
race_df.to_csv('weekly_race_to_top.csv')
```
